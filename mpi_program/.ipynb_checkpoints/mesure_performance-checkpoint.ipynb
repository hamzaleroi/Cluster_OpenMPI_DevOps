## importing the needed packages:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

## importing the dataset:

```python
url = "results.csv"
data_set = pd.read_csv("results.csv").iloc[:,[0,1,3]]
print("some data from the dataset:\n")
print(data_set[data_set['parallel-degree'] == 2].tail(5))
print(data_set[data_set['parallel-degree'] == 1].tail(5))
print("\nshape of the dataset:")
print(data_set.shape)
```

## Extracting all the possible parallelism degree:

```python
nodes_set = set(data_set.loc[:,'parallel-degree'])
nodes_set 
q1=data_set.quantile(0.25)
q3=data_set.quantile(0.75)
```

## Regression (predicting the equation):


```python
regressor = LinearRegression()
```

## Representing the graph of each topology:

```python
coefficients = []
for i in nodes_set:
    # removing outliers from the dataset and only extracting data having a special degree 
    partial_dataset = data_set.loc[(data_set['parallel-degree'] == i) \
                                   & (data_set['max-time'] < q3['max-time'] ) \
                                   &  (data_set['max-time'] > q1['max-time']),['size','max-time']]
    print("\nwith " + str(i) + " : \n")
    print(partial_dataset.tail(5))
    plt.plot(partial_dataset.iloc[:,[0]],\
             partial_dataset.iloc[:,[1]],\
             'C' + str(10 - i % 10))
    regressor.fit(partial_dataset.iloc[:,[0]],\
             partial_dataset.iloc[:,[1]])
    print("\ncoefficient:")
    print(regressor.coef_[0,0])
    dataset_linear = regressor.predict(partial_dataset.iloc[:,[0]])
    plt.plot(partial_dataset.iloc[:,[0]], dataset_linear, 'red' , linewidth=2)
    print('------------------')
    
```
